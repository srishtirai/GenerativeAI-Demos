# Exploring the Impact of Batch Normalization, Dropout, and Activation Functions on CNN Performance for CIFAR-10

This script trains multiple **CNN models on the CIFAR-10 dataset**, testing different **Batch Normalization, Layer Normalization, Dropout Rates, Number of Convolutional Layers, and Activation Functions** to analyze their impact on model performance.  

#### **Data Preprocessing**  
- Loads the **CIFAR-10 dataset** and normalizes pixel values.  
- Converts labels into **one-hot encoding** for multi-class classification.  

#### **Configurable CNN Model (`create_model`)**  
- Builds a CNN with **adjustable hyperparameters**:  
  - **Batch Normalization & Layer Normalization** for stabilizing training.  
  - **Dropout** to prevent overfitting.  
  - **Variable number of convolutional layers** for feature extraction.  
  - **ReLU, Swish, or Mish activation functions** for non-linearity.  
- Uses **AdamW optimizer** with **learning rate scheduling** to improve convergence.  

#### **Experiment Configurations**  
- Defines **eight model variations**, each modifying **dropout, normalization, layers, and activation functions** to find the best performing combination.  

#### **Training & Evaluation**  
- Each model is trained for **12 epochs** using a batch size of **64**.  
- A **learning rate scheduler** reduces the learning rate if validation loss stops improving.  
- The trained models are evaluated on the **test set**, storing accuracy and loss.  

#### **Results Visualization**  
- The results are stored in a **DataFrame** for comparison.  
- A **bar chart** is plotted to visualize the **test accuracy** of different configurations.  

This experiment helps identify the **best-performing CNN architecture** for CIFAR-10 by systematically testing **normalization techniques, dropout, depth, and activation functions**.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, optimizers
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Load CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()

# Normalize pixel values
x_train, x_test = x_train / 255.0, x_test / 255.0

# Convert labels to one-hot encoding
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)

# Function to Create an Optimized CNN Model
def create_model(batch_norm=False, dropout_rate=0.2, num_layers=3, activation="relu", layer_norm=False):
    model = models.Sequential()

    # Add Convolutional Layers (Improved Architecture)
    for i in range(num_layers):
        filters = 32 * (2**i)  # Increase filters as depth increases
        kernel_size = (5, 5) if i == 0 else (3, 3)  # Use larger filter in first layer

        model.add(layers.Conv2D(filters, kernel_size, padding='same', input_shape=(32, 32, 3) if i == 0 else ()))
        if batch_norm:
            model.add(layers.BatchNormalization())
        if layer_norm:
            model.add(layers.LayerNormalization())
        model.add(layers.Activation(activation))
        model.add(layers.MaxPooling2D((2, 2)))
        model.add(layers.Dropout(dropout_rate if i > 1 else dropout_rate / 2))  # Less dropout in early layers

    # Fully Connected Layers
    model.add(layers.Flatten())
    model.add(layers.Dense(256, activation=activation))
    model.add(layers.Dropout(dropout_rate))
    model.add(layers.Dense(10, activation='softmax'))

    # Use AdamW Optimizer for Better Convergence
    optimizer = keras.optimizers.AdamW(learning_rate=0.001, weight_decay=1e-4)

    # Compile model
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Experiment Configurations (Incorporating Improvements)
configs = [
    {"batch_norm": False, "dropout_rate": 0.2, "num_layers": 3, "activation": "relu", "layer_norm": False},  # Baseline
    {"batch_norm": True, "dropout_rate": 0.3, "num_layers": 3, "activation": "relu", "layer_norm": False},  # With Batch Norm
    {"batch_norm": True, "dropout_rate": 0.4, "num_layers": 4, "activation": "swish", "layer_norm": False},  # Deeper Model with Swish
    {"batch_norm": True, "dropout_rate": 0.4, "num_layers": 4, "activation": "swish", "layer_norm": True},  # Layer Norm instead of BN
    {"batch_norm": True, "dropout_rate": 0.3, "num_layers": 5, "activation": "swish", "layer_norm": False},  # 5-layer deep network
    {"batch_norm": True, "dropout_rate": 0.3, "num_layers": 4, "activation": "swish", "layer_norm": True},  # Batch + Layer Norm
    {"batch_norm": True, "dropout_rate": 0.3, "num_layers": 4, "activation": "mish", "layer_norm": False},  # Mish Activation
    {"batch_norm": True, "dropout_rate": 0.4, "num_layers": 5, "activation": "mish", "layer_norm": True},  # All improvements combined
]

# Train models and evaluate performance
results = []
epochs = 12
batch_size = 64

# Define model filenames
best_model_filepath = "best_cnn_model.h5"
worst_model_filepath = "worst_cnn_model.h5"

best_val_acc = 0  # Track best accuracy
worst_val_acc = 1  # Track worst accuracy

for config in configs:
    print(f"Training model with config: {config}")
    model = create_model(**config)

    # Learning Rate Scheduler
    lr_scheduler = keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, verbose=1)

    history = model.fit(
        x_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(x_test, y_test),
        callbacks=[lr_scheduler], verbose=1
    )

    # Store results
    test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
    results.append({
        "Batch Norm": config["batch_norm"],
        "Layer Norm": config["layer_norm"],
        "Dropout Rate": config["dropout_rate"],
        "Conv Layers": config["num_layers"],
        "Activation": config["activation"],
        "Test Accuracy": test_acc,
        "Test Loss": test_loss
    })

    # Save best model (highest val_accuracy)
    if test_acc > best_val_acc:
        model.save(best_model_filepath)
        best_val_acc = test_acc

    # Save worst model (lowest val_accuracy)
    if test_acc < worst_val_acc:
        model.save(worst_model_filepath)
        worst_val_acc = test_acc

# Convert results to a DataFrame and display
df_results = pd.DataFrame(results)
print("\n沒 Experiment Results:")
print(df_results)

# Plot Accuracy Comparison
plt.figure(figsize=(10, 6))
plt.barh(df_results.index, df_results["Test Accuracy"], color='skyblue')
plt.yticks(df_results.index, [f"{c['activation']} | BN: {c['batch_norm']} | LN: {c['layer_norm']} | Dropout: {c['dropout_rate']} | Layers: {c['num_layers']}" for c in configs])
plt.xlabel("Test Accuracy")
plt.title("Comparison of Different CNN Configurations")
plt.show()
```

## **Analysis of CNN Performance on CIFAR-10**

This experiment evaluates the impact of **Batch Normalization, Layer Normalization, Dropout, Convolutional Layers, and Activation Functions** on classification accuracy.

### **Effect of Batch Normalization and Dropout**
- **Batch Normalization alone does not guarantee better accuracy**. In some cases (**Config 1: 74.66%**), it performed worse than no Batch Norm (**Config 0: 77.29%**).  
- However, when **Batch Normalization is combined with Layer Normalization and Dropout (0.3-0.4)**, accuracy improved significantly (**Config 3: 81.86%**).  
- **Dropout (0.3) worked better than 0.4**, indicating that excessive dropout weakens learning.

### **Effect of Convolutional Layers**
- **Adding more layers (4-5) improved accuracy**, with the best results seen in **Config 3 (4 Layers, 81.86%)**.  
- The **5-layer models (Config 4-7) performed well (80.68%-81.00%)**, but improvements were **less significant beyond 4 layers**.  
- This suggests **a balance is needed**窶波oing too deep may lead to diminishing returns.

### **Effect of Activation Functions**
- **Swish activation (Configs 2-5) consistently outperformed ReLU (Configs 0-1)**, with **Config 3 (Swish + Layer Norm) achieving 81.86%**.  
- **Mish activation (Configs 6-7) performed well (81.00%) but was slightly less stable**.  
- **Swish + Layer Normalization (Config 3: 81.86%) provided the best results**, reinforcing the importance of normalization for deeper networks.

### **Key Takeaways**
- **Batch Normalization alone does not always improve accuracy**窶把ombining it with **Layer Normalization and Dropout (0.3)** is more effective.  
- **4-5 convolutional layers enhance feature extraction**, but **beyond 4 layers, gains are limited**.  
- **Swish activation with Layer Normalization achieved the highest accuracy (81.86%)**, making it the best-performing configuration.  


```python
from tensorflow.keras.models import load_model
import numpy as np
import matplotlib.pyplot as plt

# Define CIFAR-10 label names
cifar10_labels = [
    "Airplane", "Automobile", "Bird", "Cat", "Deer",
    "Dog", "Frog", "Horse", "Ship", "Truck"
]

# Load the best model
best_model = load_model("best_cnn_model.h5")

# Select 30 test images
num_samples = 20
sample_images = x_test[:num_samples]
actual_labels = [cifar10_labels[np.argmax(y)] for y in y_test[:num_samples]]

# Predict using the best model
best_predictions = best_model.predict(sample_images)
best_pred_labels = [cifar10_labels[np.argmax(p)] for p in best_predictions]

# Set up a grid for 20 images
fig, axes = plt.subplots(2, 10, figsize=(20, 10))

# Loop through all images and plot them
for i, ax in enumerate(axes.flat):
    ax.imshow(sample_images[i])
    ax.set_title(f"Actual: {actual_labels[i]}", fontsize=8)
    ax.axis("off")

    # Display the predicted label below the image
    ax.text(16, 38, f"Predicted: {best_pred_labels[i]}",
            fontsize=8, ha="center", color="green" if best_pred_labels[i] == actual_labels[i] else "red")

plt.tight_layout()
plt.show()
```
